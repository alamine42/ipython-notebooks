```python
true_avengers = avengers[avengers['Year'] >= 1960]
```

```python
true_avengers.shape
```

```python
def count_deaths(data):
    deaths = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
    death_count = 0
    for death in deaths:
        if data[death] == 'YES':
            death_count += 1
    return death_count
```

```python
true_avengers['Deaths'] = true_avengers.apply(count_deaths, axis=1)
```
